```python
# histogran / Distplot
fig = plt.figure(figsize=(12,8))
plt.style.use("default")
sns.distplot(df["TotalPayBenefits"],bins=50)
```

```python
# mean of totalbasepay each year
subdf = df.groupby("Year").mean()[["TotalPayBenefits"]]
subdf.reset_index(inplace=True)
subdf
```

```python
# creating lineplot
fig,ax = plt.subplots()
ax.set_title("Plot with sns",fontsize=14,fontweight="bold")
# ax.set_xticks(subdf["Year"])
sns.lineplot(x="Year",y="TotalPayBenefits",data=subdf,palette="coolwarm")
plt.show()
```

##### Jointplot

```python
# jointplots

sns.jointplot(x="OtherPay",y="TotalPay",data=df,kind="reg")
plt.show()
```

##### BarPlot 
- this is a visualization of groupby function - by default it will show mean of numeric column

```python
df["gender"].value_counts()
```

```python
# barplot
fig, ax = plt.subplots()
sns.barplot(x="Year",y="TotalPayBenefits",data=df,palette="pastel",hue="gender")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(x="Year",y="TotalPayBenefits",data=df,hue="gender",estimator=np.std)
plt.show()
```

```python
# countplot counts the no of occurances in a datacolumn - mainly catagorical data column
sns.countplot(x="gender",data=df,palette="coolwarm")
```

#### HeatMaps and clustermap

```python
# both heatmap and clustermap shows mappinf based on some matrix df which has both column and rows
# labeled like a corr matrix

# example 1
cr = df.corr()
cr
```

```python
# Heatmap
sns.heatmap(cr)
```

```python
sns.heatmap(cr,annot=True,cmap="coolwarm") # add linewidths , linecolor to add space between cells 
```

```python
# clustermap
sns.clustermap(cr,cmap="coolwarm")
```

#### Regression plots

```python
sns.lmplot(x="OtherPay",y="TotalPay",data=df, col="gender",row="Year")
```
