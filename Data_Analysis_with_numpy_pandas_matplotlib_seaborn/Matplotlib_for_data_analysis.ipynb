## Matplotlib Plots

In this topic, you will get introduced to essential parts of a Matplotlib Figure.

Figure: Whole area chosen for plotting.\
Axes: Area where data is plotted.\
Axis: Number-line like objects, which define graph limits.\
Artist: Every element on the figure is an artist.

**There are twoways for plotting charts with Matplotlib**\
**1. Functional -** with plt.plot() \
**2. Object Oriented -** using axes and fig objects


```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data 
A = np.linspace(1,5,5)
B = A**2
```

```python
%matplotlib inline 
# we dont have to type plt.show() everywhere 
```

```python
# functional method
# direclty plots on plt obj
plt.plot(A,B)
plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y label")
```

```python
plt.subplot(1,2,1)
plt.plot(A,B)

plt.subplot(1,2,2)
plt.plot(B,A,color="yellow")
```

```python
# Object Oriented method
fig = plt.figure()

ax = fig.add_subplot(111) # add_subplot(nrows, ncols, index)

ax.plot(A,B)

plt.show()
```

```python
# adjusting figsize and adding title/labels
fig = plt.figure( figsize=(8,6))
ax = fig.add_subplot(111)
ax.set( title ="liner plot", 
        xlabel ="X axis",
        ylabel = "Y axis",
      )
ax.plot(B,A, label="legend")
ax.legend()
plt.show()
```

```python
# creating subplots 
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(A,B)
ax[0].set(title="plot A")

ax[1].plot(B,A)
ax[1].set_title("plot B")

plt.tight_layout()
```

```python
# matplotlib styling 
np.random.seed(10)
day = ["mon","tues","wed","thus","fri","sat","sun"]
tempmay = np.random.randint(30,45,7);
tempjune = np.linspace(30,45,7)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(title="Temparture in may vs june",
       xlabel="Day",
       ylabel="Temp", 
      )

ax.plot(day,tempmay,label="May", color ="blue", linewidth=2,marker="o",markersize=7)
ax.plot(day,tempjune,label="June", color="purple", lw=2, marker="^",linestyle="--")
ax.legend(loc=0)
plt.show()
```

##### Scatter Plot
Scatter plot is very similar to Line Plot.

Scatter Plot is used for showing how one variable is related with another.

Scatter Plot consists of data points. If the spread of data points is linear, then two variables are highly correlated.

scatter function is used for drawing scatter plot

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
ax.scatter(days, temp, c="purple", s=[70], marker="s",label="temp",edgecolor="red")
ax.legend()
plt.show()
```

##### Bar Plot
Bar Plot is commonly used for comparing categories.

It is also used to compare categories over a short period of time.

bar and barh are used for plotting vertical and horizontal bar plots respectively

```python
#The below example plots the average sales of a company, in first three quarters of 2017.

# The code also sets the ticks on X-Axis and labels them.

quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]

fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot(111)

labels=[1,2,3]

ax.set(title="average sales of a company",
       xlabel="Quater",
       ylabel="sales",
      )
ax.set_xticks(labels)
ax.set_xticklabels(["Q1","Q2","Q3"])
ax.bar(quarters,sales_2017,label="Quarters",color="crimson",edgecolor="orange",width=0.5)
ax.legend()
plt.show()
```

```python
# bar plots with multiple values
index=[1,2,3]
x1_index=[0.8,1.8,2.8]
x2_index=[1.2,2.2,3.2]

sales_2016 = [28831, 30762, 32178]
sales_2017 = [25782, 35783, 36133]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(
    title="average sales of a company",
    xlabel="Quarter", ylabel="sales"
)

ax.bar(x1_index,sales_2016,color="wheat",width=0.4,edgecolor="black",label="2016")
ax.bar(x2_index,sales_2017,color="pink",width=0.4,edgecolor="magenta",label="2017")

ax.set_xticks(index)
ax.set_xticklabels(["Q1","Q2","Q3"])

ax.legend()
plt.show()
```

barh draws the bars horizontally as shown in above image.

height parameter is used to adjust the height of each bar.

Horizontal bar plots are used while comparing values of one category at a time.

```python
# horizontal bar charts
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales',
      xlabel='Sales (in millions)', ylabel='Quarter')
quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]
ax.barh(quarters,sales_2017,color="skyblue",height=0.6)
ax.set_yticks(index)
ax.set_yticklabels(" Q1 Q2 Q3".split())
```

### Pie Plot
Pie plot is effective in showing the proportion of categories.

It is best suited for comparing fewer categories.

In general, Pie Plot is used to highlight proportion of one or a group of categories.

```python
x=[23,57,31,7]
fig=plt.figure()
ax= fig.add_subplot(111)
ax.pie(x)
plt.show()
```

colors: Sets the colors of portions.

labels: Sets the labels of portions.

startangle: Sets the start angle at which portion drawing starts.

autopct: Sets the percentage display format of an area, covering portions.

```python
sales_2017 = [25782, 35783, 36133]
quarters = ['Q1-2017', 'Q2-2017', 'Q3-2017']
fig=plt.figure(figsize=(12,4))
ax= fig.add_subplot(111)
ax.pie(sales_2017,labels=quarters,startangle=90,autopct="%1.1f%%")
ax.set(title='Avg. Quarterly Sales')
plt.show()
```

##### Histograms

```python
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage",
      ylabel='Count', xlabel='Percentage')
ax.hist(x)
plt.show()
```

```python
import numpy as np
np.random.seed(100)
x = 60 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Distribution of Student's Percentage",
      ylabel='Proportion', xlabel='Percentage')
ax.hist(x, color='blue', bins=90, density=True)
plt.show()
```

##### BoxPlots

```python
import numpy as np
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot(x)
plt.show()
```

```python
np.random.seed(100)
x = 50 + 10*np.random.randn(1000)
y = 70 + 25*np.random.randn(1000)
z = 30 + 5*np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title="Box plot of Student's Percentage",
      xlabel='Class', ylabel='Percentage')
ax.boxplot([x, y, z], labels=['A', 'B', 'C'], notch=True, bootstrap=10000)
plt.show()
```

```python
# vertical box plots
np.random.seed(700)
A = 50 + 10*np.random.randn(1000)
B = 70 + 25*np.random.randn(1000)
C = 30 + 5*np.random.randn(1000)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(title="Box Plot", xlabel="X-axis",ylabel="Y-axis")

ax.boxplot([A,B,C],notch=True,labels="A B C".split(),bootstrap=1000,vert=False)
plt.show()
```
