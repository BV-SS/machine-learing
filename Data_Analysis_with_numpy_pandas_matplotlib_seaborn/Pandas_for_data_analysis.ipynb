#### More DataFrame methods for data filtering and sorting 
 ### Conditional selection from DF 
An important feature of pandas is conditional selection using bracket notation, very similar to numpy:

```python
df["age"] > 23
```

```python
# when cond is given inside df then only true condn rows are returned 
df[df["age"] > 23]
```

For two conditions you can use | and & with parenthesis:\
Note : python "and" and "or" will not work here as they only deal with single boolean values and these conditions rerturn a series

```python
# combining condn and displaying trageted coln
newdf = df[(df["age"] < 35) & (df["salary"] > 4500000)] 
newdf[["name","age","salary","other business"]]
```

```python
# above can be performed in one line 
df[(df["age"] < 35) & (df["salary"] > 4500000)][["name","age","salary","other business"]]
```

## Multi-Index and Index Hierarchy

Let us go over how to work with Multi-Index, first we'll create a quick example of what a Multi-Indexed DataFrame would look like:

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
mindex_df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=["A","B"])
mindex_df
```

Now let's show how to index this! For index hierarchy we use df.loc[], if this was on the columns axis, you would just use normal bracket notation df[]. Calling one level of the index returns the sub-dataframe:

```python
# How to access multi tier indexing
# using loc on outer undex returns one sub level
mindex_df.loc["G1"]
```

```python
# using loc on two levels
mindex_df.loc["G1"].loc[2]
```

```python
# indexing with columns
mindex_df.loc["G2"].loc[2]["B"]
```

using cross section function 

```python
mindex_df.xs("G1")
```

```python
# 2nd method
mindex_df.xs(["G1",2])
```

```python
mindex_df.index.names = ["Groups","Num"]
mindex_df
```

```python
# getting 1st index value from both groups
mindex_df.xs(1,level="Num") # if no name is given default indices can be used like 0,1 etc
```
